<h1 align='center' style='color:purple'>Clustering With K Means</h1>

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('income.csv')
df
```

```python
plt.scatter(df.Age,df['Income($)'])
```

```python
km = KMeans(n_clusters=3,n_init= 'auto')
km
```

```python
y_predicted = km.fit_predict(df [['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt_1 = plt.scatter(df1.Age,df1['Income($)'],color='green')
plt_2 = plt.scatter(df2.Age,df2['Income($)'],color='red')
plt_3 = plt.scatter(df3.Age,df3['Income($)'],color='black')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend([plt_1,plt_2,plt_3],['Income($)','Income($)','Income($)'])
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

```python
scaler.fit(df[['Age']])
df.Age = scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3,n_init= 'auto')
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt_1 = plt.scatter(df1.Age,df1['Income($)'],color='green')
plt_2 = plt.scatter(df2.Age,df2['Income($)'],color='red')
plt_3 = plt.scatter(df3.Age,df3['Income($)'],color='black')

plt_cen = plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = 'purple',marker ='*')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend([plt_1,plt_2,plt_3,plt_cen],['Income($) 1','Income($) 2','Income($) 3','Centroid'])
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)    #km.indertia_ = usedto calculate sum of squared error
    
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared error')
plt.plot(k_rng,sse)
```
